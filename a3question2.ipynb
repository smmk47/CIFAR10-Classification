```python
# ============================================================
# Vision Transformer (ViT), Hybrid CNN-MLP, and ResNet for CIFAR-10
# ============================================================

# --------------------
# 1. Setup and Imports
# --------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
import copy

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

import torchvision

# Set random seed for reproducibility
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f'Using device: {device}')

# -------------------------------
# 2. Data Preprocessing and Augmentation
# -------------------------------

# Define transformations for training and testing
train_transform = transforms.Compose([
    transforms.RandomCrop(32, padding=4),        # Random cropping with padding
    transforms.RandomHorizontalFlip(),           # Random horizontal flipping
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465),  # Normalize with CIFAR-10 mean and std
                         (0.2023, 0.1994, 0.2010)),
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465),
                         (0.2023, 0.1994, 0.2010)),
])

# Load the CIFAR-10 datasets
train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=train_transform)
test_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=test_transform)

# Create validation split from training data
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_subset, val_subset = random_split(train_dataset, [train_size, val_size])

# Define data loaders
batch_size = 128

train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)

print(f'Training samples: {len(train_subset)}')
print(f'Validation samples: {len(val_subset)}')
print(f'Test samples: {len(test_dataset)}')

# -----------------------------
# 2.1 Visualizing Sample Images
# -----------------------------

# Function to show images
def imshow(img, title):
    img = img / 2 + 0.5  # Unnormalize
    npimg = img.numpy()
    plt.figure(figsize=(10,10))
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    if isinstance(title, list):
        title = ', '.join(title)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Get some random training images
dataiter = iter(train_loader)
images, labels = next(dataiter)  # Corrected line

# Show images
imshow(torchvision.utils.make_grid(images[:16], nrow=4), title=[train_dataset.classes[label] for label in labels[:16]])

# -----------------------------
# 3. Vision Transformer (ViT) Implementation
# -----------------------------

class PatchEmbedding(nn.Module):
    def __init__(self, img_size=32, patch_size=4, in_channels=3, embed_dim=128):
        super(PatchEmbedding, self).__init__()
        self.patch_size = patch_size
        self.embed_dim = embed_dim
        self.num_patches = (img_size // patch_size) ** 2
        
        self.proj = nn.Conv2d(in_channels, embed_dim, kernel_size=patch_size, stride=patch_size)
    
    def forward(self, x):
        # x shape: (batch_size, in_channels, img_size, img_size)
        x = self.proj(x)  # shape: (batch_size, embed_dim, num_patches^(1/2), num_patches^(1/2))
        x = x.flatten(2)  # shape: (batch_size, embed_dim, num_patches)
        x = x.transpose(1, 2)  # shape: (batch_size, num_patches, embed_dim)
        return x

class VisionTransformer(nn.Module):
    def __init__(self, img_size=32, patch_size=4, in_channels=3, num_classes=10, 
                 embed_dim=128, depth=6, num_heads=8, mlp_dim=256, dropout=0.1):
        super(VisionTransformer, self).__init__()
        self.patch_embed = PatchEmbedding(img_size, patch_size, in_channels, embed_dim)
        num_patches = self.patch_embed.num_patches
        
        # Class token
        self.cls_token = nn.Parameter(torch.zeros(1, 1, embed_dim))
        
        # Positional encoding
        self.pos_embed = nn.Parameter(torch.zeros(1, num_patches + 1, embed_dim))
        self.dropout = nn.Dropout(dropout)
        
        # Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads, 
                                                   dim_feedforward=mlp_dim, dropout=dropout)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=depth)
        
        # Classification head
        self.mlp_head = nn.Sequential(
            nn.LayerNorm(embed_dim),
            nn.Linear(embed_dim, num_classes)
        )
        
        # Initialize weights
        self._init_weights()
    
    def _init_weights(self):
        nn.init.trunc_normal_(self.pos_embed, std=0.02)
        nn.init.trunc_normal_(self.cls_token, std=0.02)
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
            elif isinstance(m, nn.LayerNorm):
                nn.init.zeros_(m.bias)
                nn.init.ones_(m.weight)
    
    def forward(self, x):
        batch_size = x.size(0)
        x = self.patch_embed(x)  # shape: (batch_size, num_patches, embed_dim)
        
        cls_tokens = self.cls_token.expand(batch_size, -1, -1)  # shape: (batch_size, 1, embed_dim)
        x = torch.cat((cls_tokens, x), dim=1)  # shape: (batch_size, num_patches +1, embed_dim)
        
        x = x + self.pos_embed  # Add positional encoding
        x = self.dropout(x)
        
        x = x.transpose(0,1)  # Transformer expects input shape: (seq_len, batch_size, embed_dim)
        x = self.transformer(x)  # shape: (seq_len, batch_size, embed_dim)
        x = x[0]  # Take the cls token output
        
        x = self.mlp_head(x)
        return x

# Instantiate the ViT model
vit_model = VisionTransformer(
    img_size=32,
    patch_size=4,
    in_channels=3,
    num_classes=10,
    embed_dim=128,
    depth=6,
    num_heads=8,
    mlp_dim=256,
    dropout=0.1
).to(device)

print("\nVision Transformer Model:")
print(vit_model)

# -----------------------------
# 4. Hybrid CNN-MLP Architecture
# -----------------------------

class HybridCNNMLP(nn.Module):
    def __init__(self, img_size=32, patch_size=4, in_channels=3, num_classes=10, 
                 embed_dim=128, mlp_dim=256, dropout=0.1):
        super(HybridCNNMLP, self).__init__()
        self.patch_size = patch_size
        self.num_patches = (img_size // patch_size) ** 2
        
        self.cnn = nn.Sequential(
            nn.Conv2d(in_channels, embed_dim, kernel_size=patch_size, stride=patch_size),
            nn.ReLU(),
            nn.BatchNorm2d(embed_dim)
        )
        
        self.flatten = nn.Flatten(2)  # shape: (batch_size, embed_dim, num_patches)
        self.mlp = nn.Sequential(
            nn.Linear(embed_dim, mlp_dim),
            nn.ReLU(),
            nn.BatchNorm1d(mlp_dim),
            nn.Dropout(dropout),
            nn.Linear(mlp_dim, num_classes)
        )
    
    def forward(self, x):
        x = self.cnn(x)  # shape: (batch_size, embed_dim, num_patches^(1/2), num_patches^(1/2))
        x = self.flatten(x)  # shape: (batch_size, embed_dim, num_patches)
        x = x.mean(dim=2)  # Global average pooling over patches: (batch_size, embed_dim)
        x = self.mlp(x)  # Classification
        return x

# Instantiate the Hybrid CNN-MLP model
hybrid_model = HybridCNNMLP(
    img_size=32,
    patch_size=4,
    in_channels=3,
    num_classes=10,
    embed_dim=128,
    mlp_dim=256,
    dropout=0.1
).to(device)

print("\nHybrid CNN-MLP Model:")
print(hybrid_model)

# -----------------------------
# 5. ResNet with Transfer Learning
# -----------------------------

# Load a pretrained ResNet18 model
resnet_model = models.resnet18(pretrained=True)

# Freeze all layers
for param in resnet_model.parameters():
    param.requires_grad = False

# Modify the final fully connected layer
num_ftrs = resnet_model.fc.in_features
resnet_model.fc = nn.Sequential(
    nn.Linear(num_ftrs, 256),
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(256, 10)
)

resnet_model = resnet_model.to(device)

print("\nResNet18 Model with Transfer Learning:")
print(resnet_model)

# -----------------------------
# 6. Training and Hyperparameter Tuning
# -----------------------------

def train_model(model, criterion, optimizer, scheduler, num_epochs=25, early_stopping_patience=5):
    since = time.time()
    
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0
    epochs_no_improve = 0
    
    train_acc_history = []
    val_acc_history = []
    train_loss_history = []
    val_loss_history = []
    
    for epoch in range(num_epochs):
        print(f'Epoch {epoch +1}/{num_epochs}')
        print('-' * 10)
        
        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
                dataloader = train_loader
            else:
                model.eval()
                dataloader = val_loader
            
            running_loss = 0.0
            running_corrects = 0
            total_samples = 0
            
            # Iterate over data
            for inputs, labels in dataloader:
                inputs = inputs.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)
                
                # Zero the parameter gradients
                optimizer.zero_grad()
                
                # Forward
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)
                
                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()
                
                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
                total_samples += inputs.size(0)
            
            epoch_loss = running_loss / total_samples
            epoch_acc = running_corrects.double() / total_samples
            
            if phase == 'train':
                train_acc_history.append(epoch_acc.item())
                train_loss_history.append(epoch_loss)
                scheduler.step()
            else:
                val_acc_history.append(epoch_acc.item())
                val_loss_history.append(epoch_loss)
            
            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
            
            # Deep copy the model
            if phase == 'val':
                if epoch_acc > best_acc:
                    best_acc = epoch_acc
                    best_model_wts = copy.deepcopy(model.state_dict())
                    epochs_no_improve = 0
                else:
                    epochs_no_improve +=1
        
        # Early stopping
        if epochs_no_improve == early_stopping_patience:
            print('Early stopping!')
            break
        
        print()
    
    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed //60:.0f}m {time_elapsed %60:.0f}s')
    print(f'Best val Acc: {best_acc:.4f}')
    
    # Load best model weights
    model.load_state_dict(best_model_wts)
    
    history = {
        'train_acc': train_acc_history,
        'val_acc': val_acc_history,
        'train_loss': train_loss_history,
        'val_loss': val_loss_history
    }
    
    return model, history, time_elapsed

# -----------------------------
# 6.1 Training Vision Transformer (ViT)
# -----------------------------

# Define optimizer for ViT
vit_optimizer = optim.Adam(vit_model.parameters(), lr=3e-4)

# Define scheduler linked to the above optimizer
vit_scheduler = lr_scheduler.StepLR(vit_optimizer, step_size=20, gamma=0.1)

# Define hyperparameters for ViT
vit_params = {
    'model': vit_model,
    'criterion': nn.CrossEntropyLoss(),
    'optimizer': vit_optimizer,
    'scheduler': vit_scheduler,
    'num_epochs': 50,
    'early_stopping_patience': 10
}

# Train the ViT model
print("\nTraining Vision Transformer...")
vit_trained, vit_history, vit_training_time = train_model(
    model=vit_params['model'],
    criterion=vit_params['criterion'],
    optimizer=vit_params['optimizer'],
    scheduler=vit_params['scheduler'],
    num_epochs=vit_params['num_epochs'],
    early_stopping_patience=vit_params['early_stopping_patience']
)

# Save the trained ViT model
torch.save(vit_trained.state_dict(), 'vit_cifar10.pth')
print("Vision Transformer model saved as 'vit_cifar10.pth'")

# -----------------------------
# 7. Evaluation Metrics and Visualization
# -----------------------------

def evaluate_model(model, dataloader, classes):
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='macro', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1-Score: {f1:.4f}')
    
    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10,8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=classes, 
                yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
    
    # Classification Report
    print('Classification Report:')
    print(classification_report(all_labels, all_preds, target_names=classes, zero_division=0))
    
    return accuracy, precision, recall, f1, cm

# -----------------------------
# 7.1 Evaluating ViT on Test Set
# -----------------------------

# Evaluate ViT
print("\nEvaluating Vision Transformer on Test Set:")
vit_metrics = evaluate_model(vit_trained, test_loader, test_dataset.classes)

# -----------------------------
# 8. Training Hybrid CNN-MLP Architecture
# -----------------------------

# Define optimizer for Hybrid Model
hybrid_optimizer = optim.Adam(hybrid_model.parameters(), lr=1e-3)

# Define scheduler linked to the above optimizer
hybrid_scheduler = lr_scheduler.StepLR(hybrid_optimizer, step_size=20, gamma=0.1)

# Define hyperparameters for Hybrid Model
hybrid_params = {
    'model': hybrid_model,
    'criterion': nn.CrossEntropyLoss(),
    'optimizer': hybrid_optimizer,
    'scheduler': hybrid_scheduler,
    'num_epochs': 50,
    'early_stopping_patience': 10
}

# Train the Hybrid CNN-MLP model
print("\nTraining Hybrid CNN-MLP Model...")
hybrid_trained, hybrid_history, hybrid_training_time = train_model(
    model=hybrid_params['model'],
    criterion=hybrid_params['criterion'],
    optimizer=hybrid_params['optimizer'],
    scheduler=hybrid_params['scheduler'],
    num_epochs=hybrid_params['num_epochs'],
    early_stopping_patience=hybrid_params['early_stopping_patience']
)

# Save the trained Hybrid model
torch.save(hybrid_trained.state_dict(), 'hybrid_cifar10.pth')
print("Hybrid CNN-MLP model saved as 'hybrid_cifar10.pth'")

# -----------------------------
# 8.1 Evaluating Hybrid Model on Test Set
# -----------------------------

# Evaluate Hybrid Model
print("\nEvaluating Hybrid CNN-MLP on Test Set:")
hybrid_metrics = evaluate_model(hybrid_trained, test_loader, test_dataset.classes)

# -----------------------------
# 9. Training ResNet with Transfer Learning
# -----------------------------

# Optionally unfreeze some layers for fine-tuning
# Here, we freeze all layers except the last two layers of the classifier
for name, param in resnet_model.named_parameters():
    if "fc" in name:
        param.requires_grad = True
    else:
        param.requires_grad = False

# Define optimizer for ResNet (only parameters of the classifier are being optimized)
resnet_optimizer = optim.Adam(resnet_model.fc.parameters(), lr=1e-3)

# Define scheduler linked to the above optimizer
resnet_scheduler = lr_scheduler.StepLR(resnet_optimizer, step_size=10, gamma=0.1)

# Define hyperparameters for ResNet
resnet_params = {
    'model': resnet_model,
    'criterion': nn.CrossEntropyLoss(),
    'optimizer': resnet_optimizer,
    'scheduler': resnet_scheduler,
    'num_epochs': 25,
    'early_stopping_patience': 5
}

# Train the ResNet model
print("\nTraining ResNet with Transfer Learning...")
resnet_trained, resnet_history, resnet_training_time = train_model(
    model=resnet_params['model'],
    criterion=resnet_params['criterion'],
    optimizer=resnet_params['optimizer'],
    scheduler=resnet_params['scheduler'],
    num_epochs=resnet_params['num_epochs'],
    early_stopping_patience=resnet_params['early_stopping_patience']
)

# Save the trained ResNet model
torch.save(resnet_trained.state_dict(), 'resnet_cifar10.pth')
print("ResNet model saved as 'resnet_cifar10.pth'")

# -----------------------------
# 9.1 Evaluating ResNet on Test Set
# -----------------------------

# Evaluate ResNet
print("\nEvaluating ResNet on Test Set:")
resnet_metrics = evaluate_model(resnet_trained, test_loader, test_dataset.classes)

# -----------------------------
# 10. Performance Comparison
# -----------------------------

# Compile comparison metrics
comparison_df = pd.DataFrame({
    'Model': ['Vision Transformer', 'Hybrid CNN-MLP', 'ResNet'],
    'Accuracy': [vit_metrics[0], hybrid_metrics[0], resnet_metrics[0]],
    'Precision': [vit_metrics[1], hybrid_metrics[1], resnet_metrics[1]],
    'Recall': [vit_metrics[2], hybrid_metrics[2], resnet_metrics[2]],
    'F1-Score': [vit_metrics[3], hybrid_metrics[3], resnet_metrics[3]],
    'Training Time (s)': [vit_training_time, hybrid_training_time, resnet_training_time],
    # 'Memory Usage': [...],  # To be filled with actual usage if recorded
    # 'Inference Speed': [...] # To be filled with actual speeds if recorded
})

print("\nPerformance Comparison:")
print(comparison_df)

# Plotting Accuracy Comparison
plt.figure(figsize=(8,6))
sns.barplot(x='Model', y='Accuracy', data=comparison_df)
plt.title('Model Accuracy Comparison')
plt.ylim(0,1)
plt.show()

# Plotting Precision Comparison
plt.figure(figsize=(8,6))
sns.barplot(x='Model', y='Precision', data=comparison_df)
plt.title('Model Precision Comparison')
plt.ylim(0,1)
plt.show()

# Plotting Recall Comparison
plt.figure(figsize=(8,6))
sns.barplot(x='Model', y='Recall', data=comparison_df)
plt.title('Model Recall Comparison')
plt.ylim(0,1)
plt.show()

# Plotting F1-Score Comparison
plt.figure(figsize=(8,6))
sns.barplot(x='Model', y='F1-Score', data=comparison_df)
plt.title('Model F1-Score Comparison')
plt.ylim(0,1)
plt.show()

# -----------------------------
# 11. Training Curves Visualization
# -----------------------------

def plot_training_curves(history, model_name):
    epochs = range(1, len(history['train_acc']) +1)
    
    plt.figure(figsize=(14,5))
    
    # Accuracy
    plt.subplot(1,2,1)
    plt.plot(epochs, history['train_acc'], 'b-', label='Training Accuracy')
    plt.plot(epochs, history['val_acc'], 'r-', label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Loss
    plt.subplot(1,2,2)
    plt.plot(epochs, history['train_loss'], 'b-', label='Training Loss')
    plt.plot(epochs, history['val_loss'], 'r-', label='Validation Loss')
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.show()

# Plotting for ViT
print("\nTraining Curves for Vision Transformer:")
plot_training_curves(vit_history, 'Vision Transformer')

# Plotting for Hybrid Model
print("\nTraining Curves for Hybrid CNN-MLP:")
plot_training_curves(hybrid_history, 'Hybrid CNN-MLP')

# Plotting for ResNet
print("\nTraining Curves for ResNet:")
plot_training_curves(resnet_history, 'ResNet')

# -----------------------------
# 12. Saving Comparison Metrics
# -----------------------------

# Save the comparison metrics to a CSV file
comparison_df.to_csv('model_comparison.csv', index=False)
print("Model comparison metrics saved as 'model_comparison.csv'")
```

```python
# -----------------------------
# 13. Deployment and Result Visualization
# -----------------------------

# Function to visualize predictions
def visualize_predictions(model, dataloader, classes, num_images=16):
    model.eval()
    images_shown = 0
    plt.figure(figsize=(12,12))
    
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            for i in range(inputs.size(0)):
                if images_shown == num_images:
                    plt.show()
                    return
                images_shown +=1
                
                ax = plt.subplot(int(np.sqrt(num_images))+1, int(np.sqrt(num_images))+1, images_shown)
                img = inputs[i].cpu().numpy()
                # Unnormalize
                img = img * np.array([0.2023, 0.1994, 0.2010]).reshape(3,1,1) + np.array([0.4914, 0.4822, 0.4465]).reshape(3,1,1)
                img = np.clip(img, 0, 1)
                img = np.transpose(img, (1, 2, 0))
                plt.imshow(img)
                title_color = 'green' if preds[i] == labels[i] else 'red'
                plt.title(f'Pred: {classes[preds[i]]}\nTrue: {classes[labels[i]]}', color=title_color)
                plt.axis('off')
    plt.show()

# Visualize predictions for ViT
print("\nVision Transformer Predictions:")
visualize_predictions(vit_trained, test_loader, test_dataset.classes)

# Visualize predictions for Hybrid Model
print("\nHybrid CNN-MLP Predictions:")
visualize_predictions(hybrid_trained, test_loader, test_dataset.classes)

# Visualize predictions for ResNet
print("\nResNet Predictions:")
visualize_predictions(resnet_trained, test_loader, test_dataset.classes)
```
